## Vector:

#### Lets see an example 

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create a new figure and axis
fig, ax = plt.subplots()
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
```

```python
center = [0, 0]
weight1 = [2, 3]
```

## If you want to find similarity between two vector dot product can easily do that 
### Lets have an example 

```python
input_vector = [1.72 , 1.23]
weight_1 = [1.26 , 0]
weight_2 = [2.17, 0.32]
# To find the similarity between input_vector and weight 1 we will find the dot product of these two vector
# There are two way to find dot product 
#  1: 
first_indeces_mult = input_vector[0]*weight_1[0]
second_indeces_mult = input_vector[1]*weight_1[1]

dot_product_1 = first_indeces_mult + second_indeces_mult
```

```python
dot_product_1
```

### Lets create the graph so that we can visualize it 

```python
fig, ax = plt.subplots()
ax.set_xlim([0, 2.5])
ax.set_ylim([0, 2.5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
center = [0, 0]
ax.plot(center[0], center[1], 'ro')
ax.plot(weight_1[0], weight_1[1], 'bo')
ax.plot(weight_2[0], weight_2[1], 'bo')
ax.plot([center[0], weight_1[0]], [center[1], weight_1[1]], 'k-')
ax.plot([center[0], weight_2[0]], [center[1], weight_2[1]], 'k-')
```
